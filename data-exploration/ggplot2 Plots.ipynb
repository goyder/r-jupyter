# ggplot2 in R

Didjuknow? ``ggplot2`` is an implementation of the Grammar of Graphics by Leland Wilkinson.

Under this principle, the grammar of graphics represents the abstraction of graphics ideas and projects - things are broken down into verbs, nouns, adjectives, etc.

"In brief, the grammar tells us that a statistical graphic is a *mapping* from data to *aesthetic* attributes (colour, shape, size) of *geometric objects* (points, lines, bars). The plot may also contain statistical transformations of the data and is drawn on a specific coordinate system." Consequently, plots are made of 'aesthetics' and 'geoms'.

**Factors** are very important. These indicate subsets of the data.

## Fundamentals of qplot
qplot can largely be used as a replacement for plot(), but it doesn't cover all underlying work in ggplot.

```r
qplot(displ, hwy, data = mpg)  ## Hello, world!
```

```r
## Let's add a colour aesthetic.
qplot(displ, hwy, data = mpg, color = drv)
```

```r
## Let's add a geom.
qplot(displ, hwy, data = mpg, geom = c("point", "smooth"))
```

```r
## Supply a single value - get a histogram.
qplot(hwy, data = mpg, fill = drv)
```

```r
## LHS rows, RHS columns
qplot(displ, hwy, data = mpg, facets = . ~ drv)
```

```r
## LHS rows, RHS columns
qplot(hwy, data = mpg, facets = drv ~ ., binwidth = 2)
```

## In more detail: ggplot2 plots
Moving away from qplot, we might like to consider more technical ggplot2 plots.

You might consider it be composed of:
* A data frame
* aesthetic mappings (how data are mapped to color, size)
* geoms (geometric objects - points, lines, shapes, etc)
* facets (for conditional plots)
* stats (statistical transformations - binnings, quantiles, smoothing...)
* scales
* coordinate systems.

```r
library(datasets)
```

```r
colnames(airquality)
```

```r
g <- ggplot(airquality, aes(Wind, Temp))
```

```r
g + geom_point()
```

```r
g + geom_point() + geom_smooth(method = "lm")
```

```r
airquality$Month  <- factor(airquality$Month)
```

```r
g + geom_point() + facet_grid(. ~ Month) + xlab("Windy-windy-woo") + ylab("Temperature (deg F)") 
```

```r
g + geom_point() + facet_grid(. ~ Month) + xlab("Windy-windy-woo") + ylab("Temperature (deg F)") + theme_bw()
```

```r
g + 
geom_point(aes(color = Month)) + 
xlab("Windy-windy-woo") + 
ylab("Temperature (deg F)") + 
theme_bw() + 
labs(title = "Airquality dataset")
```

```r
g + 
geom_point(aes(color = Month)) + 
xlab("Windy-windy-woo") + 
ylab("Temperature (deg F)") + 
theme_bw() + 
labs(title = "Airquality dataset")
```

## Making tertiles from continuous data

```r
colnames(airquality)
```

```r
cutpoints <- quantile(airquality$Ozone, seq(0, 1, length = 4), na.rm = TRUE)
airquality$ozone_dec <- cut(airquality$Ozone, cutpoints)
```

```r
levels(airquality$ozone_dec)
```

```r
g <- ggplot(airquality, aes(x=Wind, y=Temp))
```

```r
g + geom_point() + facet_grid(. ~ ozone_dec)
```
