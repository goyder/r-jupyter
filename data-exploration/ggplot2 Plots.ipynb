# ggplot2 in R

Didjuknow? ``ggplot2`` is an implementation of the Grammar of Graphics by Leland Wilkinson.

Under this principle, the grammar of graphics represents the abstraction of graphics ideas and projects - things are broken down into verbs, nouns, adjectives, etc.

"In brief, the grammar tells us that a statistical graphic is a *mapping* from data to *aesthetic* attributes (colour, shape, size) of *geometric objects* (points, lines, bars). The plot may also contain statistical transformations of the data and is drawn on a specific coordinate system." Consequently, plots are made of 'aesthetics' and 'geoms'.

**Factors** are very important. These indicate subsets of the data.

```r
qplot(displ, hwy, data = mpg)  ## Hello, world!
```

```r
## Let's add a colour aesthetic.
qplot(displ, hwy, data = mpg, color = drv)
```

```r
## Let's add a geom.
qplot(displ, hwy, data = mpg, geom = c("point", "smooth"))
```

```r
## Supply a single value - get a histogram.
qplot(hwy, data = mpg, fill = drv)
```

```r
## LHS rows, RHS columns
qplot(displ, hwy, data = mpg, facets = . ~ drv)
```

```r
## LHS rows, RHS columns
qplot(hwy, data = mpg, facets = drv ~ ., binwidth = 2)
```
