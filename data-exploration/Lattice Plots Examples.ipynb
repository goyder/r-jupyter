# Lattice Plots Examples
Lattice is implemented with the following packages:
* `lattice`: Contains the code for calling the packages - these are typically called in a single function call.
* `grid`: Implements a lot of the low-level packages, though we rarely call it directly.

Some examples:
* `xyplot`
* `bwplot`
* `histogram`
* `stripplot` - like a boxplot, but with actual points
* `dotplot` - plot dots on 'violin strings'
* `splom` - scatterplot matrix with a fun name
* `levelplot`, `contourplot` - for plotting "image" data

## Lattice functions
Lattice functions take a formula as their first argument, something along the lines of:

`xyplot(y ~ x | f * g, data)`
`x` and `y` are the axes.

`f` and `g` are conditioning variables - they are optional, but useful. The `*` indicates there is an interaction between variables.

The `data` is the dataset to drawfrom.

## Lattice Panel Functions

We might develop lattice panels as follows:

```xyplot(y ~ x | f, panel = function(x, y, ...) {
    panel.xyplot(x, y, ...) ## call the default panel function for 'xyplot'
    panel.abline(h = median(y), lty = 2)  ## Add a horizontal line at the median
    })
    ```
