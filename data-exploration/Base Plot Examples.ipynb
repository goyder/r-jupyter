## Plotting Systems in R

The core plotting and graphics systems are encapsulated in `graphics` and `grDevices`.

`graphics` covers a lot of the fundaments like `plot,` `hist`, and `boxplot`.

`grDevices` covers the graphic devices - stuff like X11, PDF, etc.

From [here](https://stat.ethz.ch/R-manual/R-devel/library/grDevices/html/Devices.html):

The following graphics devices are currently available:

* `pdf` Write PDF graphics commands to a file
* `postscript` Writes PostScript graphics commands to a file
* `xfig` Device for XFIG graphics file format
* `bitmap` bitmap pseudo-device via Ghostscript (if available).
* `pictex` Writes TeX/PicTeX graphics commands to a file (of historical interest only)

The following devices will be functional if R was compiled to use them (they exist but will return with a warning on other systems):

* `X11` The graphics device for the X11 windowing system
* `cairo_pdf`, cairo_ps PDF and PostScript devices based on cairo graphics.
* `svg` SVG device based on cairo graphics.
* `png` PNG bitmap device
* `jpeg` JPEG bitmap device
* `bmp` BMP bitmap device
* `tiff` TIFF bitmap device
* `quartz` The graphics device for the macOS native Quartz 2d graphics system. (This is only functional on macOS where it can be used from the R.app GUI and from the command line: but it will display on the local screen even for a remote session.)

## The Three Systems of R Plotting



### Base Plotting system
General principle - an artist's palette.

You start with a blank canvas, and build up from there, adding components as you go.

```r
colnames(airquality)
```

```r
with(airquality, plot(Wind, Temp))
```

Positives: 

+ Simple, intuitive
+ It's just a series of R commands

Negatives:

- Not very translatable - each plot is fairly unique
- Pretty verbose

### Lattice System
Plot things in single function calls - e.g. `xyplot`, `bwplot`. 

`lattice` is itself composed of the `grid` package, which sort of runs independent of the "base system". 

```r
library(lattice)
```

```r
state <- data.frame(state.x77, region=state.region)
xyplot(Life.Exp ~ Income | region, data = state, layout=c(4,1))
```

Positives:
+ Margins and spacing set automatically
+ Good for putting many plots on screen
+ Good for 'conditioning' e.g. panel plots

Negatives:
- Pretty verbose when defining the function
- Annotation in plot not super intuitive
- Gotta do prep in advance to get it working

## ggplot2
Halfway between Lattice and base plot. This is often heralded as why people come to R.

```r
library(ggplot2)
```

```r
data(mpg)
qplot(displ, hwy, data = mpg)
```

Positives:
+ Splits difference between lattice and base plot

## Configuring the Base System
There are many, many things that can be customised in the base package system.

A lot of these things are documented in `par` - call this to get more deets.

```r
?par
```

But some good parameters to bear in mind:
* pch: plotting system
* lty: line type
* lwd: line width
* col: plotting color; call colors() to get examples
* xlab: character string for the x-axis label
* ylab: character string for the y-axis label

```r
colors()
```

You can get heaps of good results with `examples(points)`.

We can use the `par` function to change global parameters - some things can only be changed through the `par` function.

If we call it with the function name without arguments, we get the default values.

```r
par("bg")
```

```r
par("mar")
```

```r
par("mfrow")
```

Some core functionality in the base plotting system:
* `plot` - make a scatterplot, or at least start a plot - can specify further
* `lines` - add lines to a plot - e.g. provide X and Y data
* `points` - add points to a plot
* `text` - add text labels to a plot at specified x, y coords
* `title` - add annotation to x, y axis labels, title, subtitle, outer margin, etc
* `mtext` - add arbitrary text to margins
* `axis` - axis ticks and labels

And we can add a regression line, no problem:

```r
with(airquality, plot(Wind, Ozone, main = "Ozone and Wind in New York City", pch = 20))
model <- lm(Ozone ~ Wind, airquality)
abline(model, lwd = 2)
```
