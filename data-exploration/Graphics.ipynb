# Graphics 
A collection of examples of useful plotting techniques in R.

```r
library(datasets)
```

```r
summary(airquality)
```

```r
boxplot(airquality$Ozone, col="blue")
abline(h=100)
```

```r
hist(airquality$Ozone, col='green', breaks=30)
rug(airquality$Ozone)
```

```r
with(airquality, plot(Ozone, Wind, col=Month))
```

```r
barplot(table(airquality$Ozone), main="Ozone, my dude.")
```
